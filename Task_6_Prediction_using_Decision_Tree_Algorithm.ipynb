<h4><b>Problem statement:</b> Create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.<h4>

Dataset Link - https://bit.ly/3kXTdox

```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
#importing the dataset
df = pd.read_csv('/content/Iris (1).csv')
df.head()
```

```python
#checking for any null values
df.isna().sum()
```

```python
#looking at the shape of the dataframe
df.shape
```

```python
#dropping the ID column
df.drop(columns=['Id'],axis=1,inplace=True)
df.head()
```

```python
#plotting the values of the dataframe
plt.figure(figsize=(10,8))
sns.scatterplot(data=df)
plt.show()
```

<h3>Now Splitting dataset into Train and Test

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#splitting dataset
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

<h3>Decision Tree Algorithm

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(x_train,y_train)
```

```python
y_pred = dt_classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
matrix = confusion_matrix(y_test,y_pred)
matrix
```

```python
#visuallizing Confusion Matrix
sns.heatmap(matrix,annot=True)
```

<h3>Visualizing Decision Tree

```python
from sklearn.tree import plot_tree
fig = plt.figure(figsize=(10,20))
plot_tree(dt_classifier,feature_names=df.columns.tolist()[:-1],class_names=df.columns.tolist()[-1],filled=True)
```
