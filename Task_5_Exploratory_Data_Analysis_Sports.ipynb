```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the two dataset
match_df = pd.read_csv('/content/matches.csv')
match_df.head()
```

```python
delivery_df = pd.read_csv('/content/deliveries.csv')
delivery_df.head(10)
```

```python
#just checking the columns present in both datasets
print(match_df.columns)
print(delivery_df.columns)
```

```python
#Finding the winner of each season IPL Title
ipl_season_winners = match_df.drop_duplicates(subset=['season'],keep='last')
ipl_season_winners[['season','winner']].sort_values('season',ascending=False).reset_index(drop=True)
```

```python
ipl_season_winners[['season','winner']].sort_values('season',ascending=False).reset_index(drop=True).value_counts('winner')
```

<h4> As we can see from the above table, Mumbai Indians have won the most number of IPL titles followed by CSK with 3 Titles and KKR with 2 titles upto 2019.

```python
#Finding the number of matches won by the teams across every season of IPL
match_df[['winner']].sort_values('winner',ascending=False).value_counts('winner')
```

<h4>Across all the IPL seasons, Mumbai Indians have won the most IPL matches followed by CSK and KKR.

```python
#One thing we can notice is that Their are two entries of the same team because of different spelling
match_df['winner'].replace('Rising Pune Supergiants','Rising Pune Supergiant',inplace=True)
match_winners_total = match_df[['winner']].sort_values('winner',ascending=False).value_counts('winner')
match_winners_total
```

```python
#plotting number of matchs won by teams
plt.figure(figsize=(15,10))
sns.countplot(match_df['winner'])
plt.title('Matches Won by Team')
plt.xlabel('Teams ->')
plt.ylabel('Wins ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#Finding the most rus scored by players
player_runs = delivery_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
player_runs[:10]
```

```python
plt.figure(figsize=(15,10))
sns.barplot(player_runs[:10].index,player_runs[:10])
plt.title('Most Runs ')
plt.xlabel('Batsmen ->')
plt.ylabel('Runs ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#Now finding the bowlers with most number of wickets
delivery_df['Out'] = [0 if pd.isna(x) else 1 for x in delivery_df['player_dismissed']]
```

```python
delivery_df
```

```python
wickets_players = delivery_df.groupby('bowler')['Out'].sum().sort_values(ascending=False)
wickets_players
```

```python
plt.figure(figsize=(15,10))
sns.barplot(wickets_players[:10].index,wickets_players[:10])
plt.title('Most Wickets')
plt.xlabel('Bowlers ->')
plt.ylabel('Wickets ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#finding the most number of fours scored per team and per player
delivery_df.dtypes
```

```python
delivery_df['batting_team'].replace('Rising Pune Supergiants','Rising Pune Supergiant',inplace=True)
delivery_df['bowling_team'].replace('Rising Pune Supergiants','Rising Pune Supergiant',inplace=True)
fours_df = delivery_df[delivery_df['total_runs']==4]
fours_scored_team = fours_df.groupby('batting_team')['total_runs'].count().sort_values(ascending=False)
fours_scored_team
```

```python
#Fours scored per team
plt.figure(figsize=(15,10))
sns.barplot(fours_scored_team.index,fours_scored_team)
plt.title('Fours Scored per Team')
plt.xlabel('Teams ->')
plt.ylabel('Fours ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#Fours Scored per Player
fours_scored_player = fours_df.groupby('batsman')['total_runs'].count().sort_values(ascending=False)
fours_scored_player
```

```python
#Most Fours Scored by Players
plt.figure(figsize=(15,10))
sns.barplot(fours_scored_player[:10].index,fours_scored_player[:10])
plt.title('Fours Scored per Player')
plt.xlabel('Players ->')
plt.ylabel('Fours ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#Finding the most number of sixes scored by each team and player
six_df = delivery_df[delivery_df['total_runs']==6]
six_scored_team = six_df.groupby('batting_team')['total_runs'].count().sort_values(ascending=False)
six_scored_team
```

```python
#Six scored per team
plt.figure(figsize=(15,10))
sns.barplot(six_scored_team.index,six_scored_team)
plt.title('Six Scored per Team')
plt.xlabel('Teams ->')
plt.ylabel('Six ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#Six Scored per Player
six_scored_player = six_df.groupby('batsman')['total_runs'].count().sort_values(ascending=False)
six_scored_player
```

```python
#Most Six Scored by Players
plt.figure(figsize=(15,10))
sns.barplot(six_scored_player[:10].index,six_scored_player[:10])
plt.title('Six Scored per Player')
plt.xlabel('Players ->')
plt.ylabel('Six ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#Finding the players getting the most Player of the Match
player_of_match = match_df['player_of_match'].value_counts()
player_of_match
```

```python
#Player of the Match
plt.figure(figsize=(15,10))
sns.barplot(player_of_match[:10].index,player_of_match[:10])
plt.title('Most Player of Match Awards')
plt.xlabel('Players ->')
plt.ylabel('Total Awards ->')
plt.xticks(rotation=60)
plt.show()
```

```python
#finding out the matches won based on toss decision
match_df['Won'] = np.where(match_df['win_by_runs'],'Bat First','Bowl First')
```

```python
#plotting pie chart of Win based on Toss Decision
wins = match_df.Won.value_counts()
labels = np.array(wins.index)
size = wins.values
plt.figure(figsize=(10,8))
plt.pie(size,labels=labels,autopct='%1.1f%%',startangle=90)
plt.title('Match Result based on Toss (Win %)',fontsize=15)
plt.show()
```

```python
match_df['venue'].unique()
```

```python
plt.figure(figsize = (20,15))
sns.countplot('venue',hue='Won',data=match_df)
plt.title("Wins based on Toss decision at each Venue",fontsize=20)
plt.xlabel("Stadium",fontsize=15)
plt.ylabel("Wins",fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

<h4>From the graph above, we can see that M Chinnaswamy Stadium, Wankhede Stadium and Eden Gardens teams which Bowl First have more chances of winning then teams Batting First. This is a trend which we can observe pretty much at every venue.

```python
# we will plot graph on Numbers of matches won by Toss result
plt.figure(figsize = (20,12))
sns.countplot('season',hue='toss_decision',data=match_df)
plt.title("Numbers of matches won by Toss result ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

<h4>As we can observe from the graph given above, especially over the past few seasons, we can see that teams selecting Feilding First win majority of the matches. Thus giving us an observation that teams selecting to Feild First (or Bowl First) tend to win majority of the matches in the IPL. (Especailly after 2016 season)

```python
#plotting the most way dismissals have taken place
plt.figure(figsize=(20,12))
ax=sns.countplot(delivery_df['dismissal_kind'])
plt.title("Dismissals in IPL")
plt.xlabel("Dismissals kind")
plt.ylabel("count")
plt.xticks(rotation=90)
plt.show()
```
